Por otro lado, grid activa o desactiva la cuadrícula en el gráfico. Si el valor es True, la cuadrícula se activa. La cuadrícula es una serie de líneas horizontales y verticales que pueden ayudar a visualizar mejor los datos.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

class DataAnalysis:
    def __init__(self, data):
        self.data = data
    
    def reshape_data(self, new_shape):
        self.data = self.data.reshape(new_shape)
    
    def convert_to_dataframe(self, columns):
        self.data = pd.DataFrame(self.data, columns=columns)
    
    def plot_signal(self, title="", xlabel="", ylabel="", legend=True, grid=True):
        plt.plot(self.data)
        plt.title(title)
        plt.xlabel(xlabel)
        plt.ylabel(ylabel)
        if legend:
            plt.legend()
        if grid:
            plt.grid()
    
    def plot_histogram(self, title="", xlabel="", ylabel="", bins=10, legend=True, grid=True):
        plt.hist(self.data, bins=bins)
        plt.title(title)
        plt.xlabel(xlabel)
        plt.ylabel(ylabel)
        if legend:
            plt.legend()
        if grid:
            plt.grid()
    
    def plot_stem(self, title="", xlabel="", ylabel="", legend=True, grid=True):
        plt.stem(self.data)
        plt.title(title)
        plt.xlabel(xlabel)
        plt.ylabel(ylabel)
        if legend:
            plt.legend()
        if grid:
            plt.grid()
    
    def plot_bar(self, title="", xlabel="", ylabel="", legend=True, grid=True):
        plt.bar(range(len(self.data)), self.data)
        plt.title(title)
        plt.xlabel(xlabel)
        plt.ylabel(ylabel)
        if legend:
            plt.legend()
        if grid:
            plt.grid()
    
    def plot_pie(self, title="", legend=True, grid=True):
        plt.pie(self.data, labels=self.data.index)
        plt.title(title)
        if legend:
            plt.legend()
        if grid:
            plt.grid()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

class MyDataAnalyzer:
    def __init__(self, data):
        self.data = data
        plt.ion()
        
    def to_dataframe(self):
        df = pd.DataFrame(self.data)
        return df
        
    def sum(self, axis=0):
        return np.sum(self.data, axis=axis)
        
    def subtract(self, axis=0):
        return np.subtract(self.data, np.mean(self.data, axis=axis))
        
    def multiply(self, axis=0):
        return np.prod(self.data, axis=axis)
        
    def divide(self, axis=0):
        return np.divide(self.data, np.std(self.data, axis=axis))
        
    def average(self, axis=0):
        return np.mean(self.data, axis=axis)
        
    def std_deviation(self, axis=0):
        return np.std(self.data, axis=axis)
        
    def plot_signal(self, title="", x_label="", y_label="", legend=False, grid=False):
        plt.figure()
        plt.title(title)
        plt.xlabel(x_label)
        plt.ylabel(y_label)
        plt.plot(self.data)
        
        if legend:
            plt.legend()
            
        if grid:
            plt.grid()
            
        plt.show()
        
    def plot_histogram(self, title="", x_label="", y_label="", legend=False, grid=False):
        plt.figure()
        plt.title(title)
        plt.xlabel(x_label)
        plt.ylabel(y_label)
        plt.hist(self.data)
        
        if legend:
            plt.legend()
            
        if grid:
            plt.grid()
            
        plt.show()
        
    def plot_stem(self, title="", x_label="", y_label="", legend=False, grid=False):
        plt.figure()
        plt.title(title)
        plt.xlabel(x_label)
        plt.ylabel(y_label)
        plt.stem(self.data)
        
        if legend:
            plt.legend()
            
        if grid:
            plt.grid()
            
        plt.show()
        
    def plot_bar(self, title="", x_label="", y_label="", legend=False, grid=False):
        plt.figure()
        plt.title(title)
        plt.xlabel(x_label)
        plt.ylabel(y_label)
        plt.bar(range(len(self.data)), self.data)
        
        if legend:
            plt.legend()
            
        if grid:
            plt.grid()
            
        plt.show()
        
    def plot_pie(self, title="", legend=False, grid=False):
        plt.figure()
        plt.title(title)
        plt.pie(self.data)
        
        if legend:
            plt.legend()
            
        if grid:
            plt.grid()
            
        plt.show()

def main():
    # Example usage
    data = np.random.rand(288, 160, 250)
    analyzer = MyDataAnalyzer(data.reshape(288, 160*250))
    df = analyzer.to_dataframe()
    s = analyzer.sum(axis=0)
    sub = analyzer.subtract(axis=1)
    mul = analyzer.multiply(axis=2)
    div = analyzer.divide(axis=0)
    avg = analyzer.average(axis=1)
    std = analyzer.std_deviation(axis=2)
```
